```python
file = '/Users/akerr/Desktop/urban-doodle/Columns.txt'
f=open(file,"r")
lines=f.readlines()
result=[]
for p in lines:
    #print(x)
    result.append(p.split())
f.close()
```

```python
positions = []
vel = []
acc = []
kinetic = []
potential = []
total_energy = []
for i in range(len(result)-1):
    positions.append(float(result[i][1]))
    vel.append(float(result[i][2]))
    acc.append(float(result[i][3]))
    kinetic.append(float(result[i][4]))
    potential.append(float(result[i][5]))
    total_energy.append(float(result[i][6]))
    
```

```python
plt.title("Positions")
plt.xlabel("Time Steps")
plt.ylabel("Displacement")
plt.plot(positions, color = 'sienna')
```

```python
plt.title("Velocity")
plt.xlabel("Seconds")
plt.ylabel("Meters")
plt.plot(vel, color = 'darkgoldenrod')
```

```python
plt.plot(acc, color = 'darkolivegreen')
plt.title("Accelleration")
plt.xlabel("Time Steps")
plt.ylabel("Mass")
```
