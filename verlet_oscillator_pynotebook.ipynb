```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
#lets sample harmonic oscillator for 25 seconds and let dt = .1 seconds
#N is the number of steps
#let t be in seconds
#let x be in meters
#let m be in kg
#k must be in kg/s**2
N = 250
k = 1
m = 1
x = 0
v = 1
dt = .1
#PE = 1/2*k*x**2
#KE = 1/2*m*v**2
#a = F/m = -kx/m
```

```python
#update position N times
position = []
time = []
velocity = []
PE_l = []
KE_l = []
TE_l =[]
x_o = x
x = x + dt * v
for t in range(N):
    PE = 1/2*k*x**2
    KE = 1/2*m*v**2
    TE = PE + KE
    PE_l.append(PE)
    KE_l.append(KE)
    TE_l.append(TE)
    velocity.append(v)
    position.append(x)
    time.append(t * dt)
    #since F = -k/x
    a     = -k * x / m
    #then update position in x_n using verlet
    x_n = 2.0 * x - x_o + (dt*dt) * a
    #likewise update velocity
    v     = (x_n - x)/(dt)
    #current step becomes previous
    x_o = x
    #new step becomes current
    x     = x_n
    
#iterate N times
```

```python
plt.scatter(time, position ,color = 'blue', s = .1, label = "position")
plt.scatter(time, velocity ,color = 'orange', s = .1, label = "velocity")
plt.legend(loc="upper left")
plt.xlabel('time (s)')
```
