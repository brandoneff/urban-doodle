```python
import numpy as np
import os
from matplotlib import pplot as plt 
plt.style.use('ggplot')
```

```python
k = 4.20 # constant
x_0 = 0 # equilibrium (initial position)
x = 1 # displacement of the spring
v = 0 # velocity
F = -k * (x-x_0) # Hooke's Law
m = 1 # mass 
a = F / m # acceleration
w = 1 # Frequency

N = 1000 # number of steps
dt = .01 # time step

positions_py = np.zeros(N)
velocities = np.zeros(N)

for i in range(N):
    positions_py[i] = x
    velocities[i] = v
    x = x + (v * dt + ((1 / 2) * a * (dt ** 2)))
    F = -k * (x-x_0)
    a_next = (F / m)
    v = v + (dt * (1 / 2) * (a+a_next))
    a = a_next
```

```python
plt.title("Positions Test Data: Python")
plt.xlabel("Time Steps")
plt.ylabel("Displacement")
plt.plot(positions_py, color = 'maroon')
```
