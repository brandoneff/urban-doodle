```python
import numpy as np
import os
from matplotlib import pyplot as plt 
plt.style.use('ggplot')
```

```python
k = 4.20 # constant kg/s**2
x_0 = 0 # equilibrium (initial position)
x = 1 # displacement of the spring
v = 0 # velocity m/s
F = -k * (x-x_0) # Hooke's Law
m = 1 # mass kg
a = F / m # acceleration m/s**2
w = 1 # Frequency

N = 100 # number of steps
dt = 1 # time step
```

```python
# From Python for Verlet: TEST 3

v_0 = 0.0   #initial velocity
dv = (k * x) / m # acceleration

position_e = np.zeros(N) # initialization
velocity_e = np.zeros(N)
kinetic_energy = np.zeros(N)
potential_energy = np.zeros(N)
total_energy = np.zeros(N)

for i in range(N): # for loop for each time step
    
    position_e[i] = x + dt * v # denotes the position at a specific time step based on the (old) velocity
    velocity_e[i] = v + dv * dt # denotes the velocity at a specific time step based on the (old) acceleration
    
    x = (x + v * dt) # updates the position
    v = (v - (k / m) * x * dt) # updates the velocity
    
    kinetic_energy[i] = 0.5 * m * (v ** 2) # calculates the kinetic energy for the given time step
    potential_energy[i] = 0.5 * k * (x ** 2) # calculates the potential energy for the given time step 
    total_energy[i] = kinetic_energy[i] + potential_energy[i] # calculates the total energy for the given time step
```
