```python
def distance(particle1, particle2):
    """Calculates the distance between two particles given their coordinates.
        INPUT
    particle1 = coordinates (three dimensions) of the first particle
    particle2 = coordinates (three dimensions) of the second particle
        OUTPUT
    calc_distance = calculated distance (in angstroms)"""
    
    del_x = particle1[0] - particle2[0] # calculates delta x
    del_y = particle1[1] - particle2[1] # delta y
    del_z = particle1[2] - particle2[2] # and delta z
    dis_vector = np.array((del_x, del_y, del_z)) # forms vector like Dr Heyden asked this will be used for PBC
    
    calc_distance = np.sqrt(np.power(del_x, 2) + np.power(del_y, 2) + np.power(del_z, 2)) # distance formula
    # note that this is going to return 1 number, and only works for when two particles are entered. 
    # TODO how can i call this in a for loop so that given more than two particles i can calculate the distance
    # between them all?
    
    return calc_distance #, dis_vector NOTE this is the vector that Dr Heyden was talking about
```

```python
def lj_potential(r, e, sig): # in (kJ / mol) 
    
    """Calculates the Lennard Jones Potential Energy given the distance between particles.
        INPUT
    r = distance formula
    e = well-depth (how strongly the particles attract one another)
    sig = the distance at which the potential energy is zero
        OUTPUT
    PE = potential energy of the two particles"""
    
    # e = the depth and measure of how strongly the particles 
    # attract one another/equilibrium potential energy
    # r = the distance at which intermolecular potential is zero
    # r = distance in angstroms
    # e = kJ/mol = (kg * meters ** 2) / ((s ** 2) * mol)
    # sig = van der waals radius in angstroms
    
    PE = 4 * e * ((sig / r) ** 12 - (sig / r) ** 6) # potential accounts for the difference
    #between two nonbonding molecules based on their attractive/repulsive forces
    # sig / r ** 12 is the repulsive force
    # sig / r ** 6 is the attractive force
    # all multiplied by the well depth (how strongly the particles attract one another)

    #print(lj_poten)
    return PE

#print(lj_potential(r, e, sig)) 
# test: particle 1 is at [0, 0, 0] and particle 2 is at [3.816, 0, 0]. 
# we should expect the potential energy to be very close to -1 at this point
# based on my test graph (see desmos) where e = 0.997, sig = 3.40, and thus r = 3.816
```

```python
def lj_force(r, e, sig): # in (kJ / (mol * angstroms)
    
    """Calculates the Lennard Jones Force given the distance between some particles.
        INPUT
    r = distance formula (in angstroms)
    e = well-depth (how stormgly the particles attract)
    sig = the distance at which potential energy is zero
        OUTPUT
    F = force"""
    
    # force magnitude: the negative gradient of potential energy
    # distance formula (angstroms)
    # e = equlibrium potential energy in kJ/mol
    # sig = van der waals radius in angstroms
    F = -4 * e * ((-12 * np.power(sig, 12) * np.power(r, -13)) + (6 * np.power(sig, 6) * np.power(r, -7)))

    return F # magnitude of the force
#print(lj_force(r, e, sig))

# test: since the minimum of potential energy was at a distance of appx. 3.816 angstroms
# we should expect the force to be equal to 0 at that point and be changing from 
# negative to positive at this point, which can be observed on the graph
# (as well as on my test in desmos)
```

```python
def acceleration(particle_positions):
    
    """Using an array which contains the x, y, and z coordinates of a given amount of particles
    the acceleration is calculated:
        INPUT
    particle_positions = an array containing coordinates
        OUTPUT
    acc = the accelerations in an array of the same dimensions as particle_positions"""
    
    acc_matrix = np.zeros((np.size(particle_positions, 0), np.size(particle_positions, 1)))
    # forms an empty matrix where acceleration values will be entered
    # the first list will be the acceleration on particle 1 as a result of particle 2
    # the second list will be the accacceleration on particle 2 as a result of particle 1 and so on
    
    # now we will begin the calculation of acceleration on the particles
    for i in range(len(particle_positions) - 1): # accessing the first particle
        for j in range(i + 1, len(particle_positions)): # accessing the second particle
            # we need to access both particles in order to calculate the acceleration
            
            vector = np.array(particle_positions[j] - particle_positions[i])
            
            # forms a vector based on the difference between the coordinates of the indexed particles
            # the point of the loop is so that if we had more than 2 particles we would be able to
            # calculate their acceleration
            
            r_x = vector[0] # here we're accessing each element of the vector
            r_y = vector[1] # we use these values to calculate the distance
            r_z = vector[2] # afterward we can form a normalized force vector
            
            rmag = distance(particle_positions[i], particle_positions[j])
            #rmag = np.sqrt(np.power(r_x, 2) + np.power(r_y, 2) + np.power(r_z, 2))
            # i changed rmag so that instead of manually calculating the distance i can do it via the distance function
            
            force_scalar = lj_force(rmag, 0.997, 3.4) # calculates the force magnitude
            # force_scalar is 1D. we then multiply it to the x,y, and z factors to become 3D
            
            force_x = force_scalar * (r_x / rmag) # normalizing the r_x, r_y, and r_z vectors
            force_y = force_scalar * (r_y / rmag) # then multiplying by the force magnitude
            force_z = force_scalar * (r_z / rmag) # this is how we form the force vector 
            
            forces_xyz = np.array([force_x, force_y, force_z]) # maybe if i make an array of them all i can index & add?
            print(forces_xyz)
            
            
            
            acc_matrix[i][0] = -(force_x / mass) # calculates the acceleration on particle i from j
            acc_matrix[j][0] = (force_x / mass) # calculates the acceleration on particle j from i
            # indexing at [i, 0] and [j, 0] dictates which particle the acceleration is assigned to
            
            acc_matrix[i][1] = -(force_y / mass)
            acc_matrix[j][1] = (force_y / mass)
            
            acc_matrix[i][2] = -(force_z / mass)
            acc_matrix[j][2] = (force_z / mass)
             
             
    return acc_matrix
acceleration(grid)
```

```python
# VERLET PLANNING NEW

dt = 0.01 # time step
vel = 0 # velocity (meters / s)
acc = acceleration(grid) # calling the acceleration function
```

```python
def one_step(pos, vel_i, time_step, accels): 
    
    """Calculates the change in position and velocity for one time step.
        INPUT
    pos = initial position of particles
    vel_i = initial velocity (I chose 0)
    time_step = a given change in seconds
    accels = initial acceleration of the particles
        OUTPUT
    new_pos = the updated position of both particles after one time step change
    new_vel = the updated velocities of both particles after one time step change"""
    
    for ele in pos: # accesses each particle
        
        old_pos = pos # sets the initial position
        acc_next = acceleration(pos) # sets acc_next variable and calls the acceleration function
        #print(acc_next)
        new_pos = old_pos + (vel_i * time_step) + ((1 / 2 * accels) * (time_step ** 2)) # verlet algorithm for position
        new_vel = vel_i + (((acc_next + accels) / 2) * time_step) # verlet algorithm for velocity
        #time += time_step # changes the time step
        
    return new_pos, new_vel, acc_next # returns updated position and velocity

# one_step should return as follows:
# an array which contains the updated positions for particles one and two as a 2x3 matrix
# an array which contains the updated velocities for particles one and two as a 2x3 matrix
# expecting a change in position by +- 0.68501 but am getting 4x that value, why?
```

```python
# VERLET FOR ALL TIME STEPS
N = 10000

pos_data = []
vel_data = []
acc_data = []

temp_pos, temp_vel, temp_acc = one_step(grid, vel, dt, acc)
for i in range(N):
    temp_pos, temp_vel, temp_acc = one_step(temp_pos, temp_vel, dt, temp_acc)
    pos_data.append(temp_pos)
    
```
