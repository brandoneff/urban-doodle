```python
import numpy as np
import os
from matplotlib import pyplot as plt 
plt.style.use('ggplot')
```

```python
k = 4.20 # constant
x_0 = 0 # equilibrium (initial position) meters
x = 1 # displacement of the spring meters (current position)
v = 0 # velocity m/s
F = -k * (x-x_0) # Hooke's Law 
m = 1 # mass kg
a = F / m # acceleration m/s**2
#w = 1 # Frequency

N = 100 # number of steps
dt = .1 # time step

positions_py = np.zeros(N) # initialization
velocities = np.zeros(N)
kinetic_e = np.zeros(N)
potential_e = np.zeros(N)
total_e = np.zeros(N)

for i in range(N): # for loop for each time step
    positions_py[i] = x # sets the position at the current time step
    velocities[i] = v # sets the velocity at the current time step
    x = x + (v * dt + ((1 / 2) * a * (dt ** 2))) # updates the position
    a_next = (F / m) # defines the new acceleration
    v = v + (dt * (1 / 2) * (a+a_next)) # updates the velocity
    # notice that the Verlet uses the average of the acceleration and that Euler just uses the
    # old acceleration to update the new velocity, thus creating an error over time
    a = a_next # updates the acceleration
    
    kinetic_e[i] = 0.5 * m * (v ** 2) # calculates the kinetic energy at the given time step 
    potential_e[i] = 0.5 * k * (x ** 2) # calculates the potential energy at the given time step
    total_e[i] = kinetic_e[i] + potential_e[i] # calculates the total energy at the given time step
```

```python
# Subplots for position, velocity and total energy
#leftmost
plt.subplots(1,3,figsize=(15,5))
plt.suptitle("Verlet", fontsize=20)

plt.subplot(1, 3, 1)
plt.plot(positions_py, color = 'maroon')
plt.title('Position')
plt.xlabel('Time Steps')
plt.ylabel('Displacement')

#middle
plt.subplot(1, 3, 2)
plt.plot(velocities, color = 'darkgoldenrod')
plt.title('Velocity')
plt.xlabel('Time (s)')
plt.ylabel('Meters')

#rightmost figure
plt.subplot(1, 3, 3)
plt.plot(total_e, color = 'black')
plt.title('Total Energy')
plt.xlabel('Time (s)')
plt.ylabel('Joules')


plt.savefig("/Users/akerr/Desktop/my_first_figure.png")
```
