```python
import numpy as np
from matplotlib import pyplot as plt 
```

```python
k = 10 # constant
x_0 = 0 # equilibrium
x = 0 
v = 3 # velocity
F = -k * (x-x_0) # Hooke's Law
m = 1 # mass 
a = F / m # acceleration

N = 500 # number of steps
dt = .05 # time

positions = np.zeros(N)
velocities = np.zeros(N)

for i in range(N):
    positions[i] = x
    velocities[i] = v
    x = x + (v * dt + ((1 / 2) * a * (dt ** 2)))
    F = -k * (x-x_0)
    a_next = (F / m)
    v = v + (dt * (1 / 2) * (a+a_next))
    a = a_next
```

```python
plt.plot(positions)
```
